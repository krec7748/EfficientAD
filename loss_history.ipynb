```python
import torch
import matplotlib.pyplot as plt
```

```python
loss_history_path = "./runs/juice_bottle/train_01/loss_history.pt"
loss_history = torch.load(loss_history_path)
```

```python
train_loss = loss_history["train_loss"]
val_loss = loss_history["val_loss"]
auc_score = loss_history["auc"]

epochs = range(1, len(train_loss) + 1)

# Visualization
plt.figure(figsize=(12, 6))

# Loss
plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, label="Train Loss", marker='o')
plt.plot(epochs, val_loss, label="Val Loss", marker='o')
plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid()

# AUC
plt.subplot(1, 2, 2)
plt.plot(epochs, auc_score, label="AUC Score", color='green', marker='o')
plt.title("AUC Score vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("AUC Score")
plt.legend()
plt.grid()

# Show
plt.tight_layout()
plt.show()
```
